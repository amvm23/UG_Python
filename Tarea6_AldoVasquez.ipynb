## Diferencia entre array y matrix en NumPy ##

Básicamente el objeto matrix está contenido en el objeto array, ya que este último es un tensor que puede ir desde 1 a n dimensiones, por lo que hereda todas las propiedades de array

La facilidad de uso es que cuando estemos multiplicando o realizando cualquier otra operación, no nos debemos preocupar por que se puedan operar ya que todos los objetos Matrix tienen la mimsma dimensión




```python
# Bellman Equation

import numpy as np
R = np.array([[1],
             [2],
             [3]])

y = 5

P = np.array([[7,8,5],
              [9,10,3],
              [11,12,2]])

V = np.array([[2],
             [2],
             [3]])

v = R + y*np.matmul(P,V)


print(v)
```
