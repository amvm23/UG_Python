GIT
Git es un software para el control de versiones, es ampliamente utilizado en el mundo del desarrollo de software, más sin embargo, su utilidad no se limita solo para desarrolladores de software

Git nos permite crear un repositorio con todos los archivos del proyecto y podemos ir sacando como fotos instantáneas a determinado momento cuando nosotros lo necesitemos.  

Lo interesante de git es que si queremos regresar a una versión anterior lo podemos realizar, así mismo si queremos trabajar en una versión alternativa lo podemos hacer creando un branch, con esta nueva versión, posteriormente podemos hacer un merge con la versión original o sustituir la última versión con esta nueva versión del branch que creamos

Pero talvez lo más importante de git es que no solo nos sirve para llevar el contrl de versiones de nuestro propio repositorio, es decir, podemos conectarnos a un repositorio compartido y de esta manera estar trabajando concurrentemente con otros desarrolladores sobre el mismo proyecto, cada quién en una parte del proyecto y luego unir todo en un solo proyecto.  Para esto debemos tener un servidor compartido al cuál enviaremos nuestro propio repositorio local

Vamos a diferenciar 3 áreas con las que trabaja Git:

Repositorio: Es el área del disco duro que contiene los archivos de nuestro proyecto bajo el control de versiones de Git
Stage: El stage area es un área en memoria donde Git tendrá los archivos que necesitan confirmarse al repositorio

Iniciar
Lo primero que debemos hacer es indicarle a git la carpeta donde está ubicado nuestro proyecto, esto lo logramos con la instrucción [git init]

Agregar los archivos
Lo siguiente que debemos hacer es indicarle a git los archivos a los que Git debe darle un tracking para verificar si hay cambios que necesitan ser enviados posteriormente al repositorio

Confirmar
Una vez los archivos fueron agregados y Git tiene conocimiento de ellos, Git tiene la capacidad de decirnos si han sufrido algún cambio con respecto a la versión que tenemos almacenada en el Repositorio

Status
Un comando muy útil es [git status] el cuál nos indicar los archivos que se han editado y están listos para ser confirmados

Crear un branch

Como mencioné arriba, es posible crear un proyecto alterno (sobre el mismo proyecto), esto lo hacemos creando un branch, lo que nos permitirá hacer cambios, sin caerle encima a la versión original de nuestro proyecto.  Nosotros podemos decidir si trabajamos sobre la versión originall [root] o sobre la nueva branch que creamos

Buscar diferencias
Un comando muy útil es [git diff] el cuál nos ayuda a encontrar las diferencias entre 2 archivos de nuestro proyecto, esto es útil si por ejemplo cometimos un error y queremos comparar nuestro código actual contra la última versión que no contenía el error, de esta manera podemos ver que fué lo que cambiamos


Confirmaciones al remoto
Para enviar la última confirmación de nuestro [root] utilizamos el comando [git push]
