GIT
Git es un software para el control de versiones, es ampliamente utilizado en el mundo del desarrollo de software, ms sin embargo, su utilidad no se limita solo para desarrolladores de software

Git nos permite crear un repositorio con todos los archivos del proyecto y podemos ir sacando como fotos instantneas a determinado momento cuando nosotros lo necesitemos.  

Lo interesante de git es que si queremos regresar a una versin anterior lo podemos realizar, as mismo si queremos trabajar en una versin alternativa lo podemos hacer creando un branch, con esta nueva versin, posteriormente podemos hacer un merge con la versin original o sustituir la ltima versin con esta nueva versin del branch que creamos

Pero talvez lo ms importante de git es que no solo nos sirve para llevar el contrl de versiones de nuestro propio repositorio, es decir, podemos conectarnos a un repositorio compartido y de esta manera estar trabajando concurrentemente con otros desarrolladores sobre el mismo proyecto, cada quin en una parte del proyecto y luego unir todo en un solo proyecto.  Para esto debemos tener un servidor compartido al cul enviaremos nuestro propio repositorio local

Vamos a diferenciar 3 reas con las que trabaja Git:

Repositorio: Es el rea del disco duro que contiene los archivos de nuestro proyecto bajo el control de versiones de Git
Stage: El stage area es un rea en memoria donde Git tendr los archivos que necesitan confirmarse al repositorio

Iniciar
Lo primero que debemos hacer es indicarle a git la carpeta donde est ubicado nuestro proyecto, esto lo logramos con la instruccin [git init]

Agregar los archivos
Lo siguiente que debemos hacer es indicarle a git los archivos a los que Git debe darle un tracking para verificar si hay cambios que necesitan ser enviados posteriormente al repositorio

Confirmar
Una vez los archivos fueron agregados y Git tiene conocimiento de ellos, Git tiene la capacidad de decirnos si han sufrido algn cambio con respecto a la versin que tenemos almacenada en el Repositorio

Status
Un comando muy til es [git status] el cul nos indicar los archivos que se han editado y estn listos para ser confirmados

Crear un branch

Como mencion arriba, es posible crear un proyecto alterno (sobre el mismo proyecto), esto lo hacemos creando un branch, lo que nos permitir hacer cambios, sin caerle encima a la versin original de nuestro proyecto.  Nosotros podemos decidir si trabajamos sobre la versin originall [root] o sobre la nueva branch que creamos

Buscar diferencias
Un comando muy til es [git diff] el cul nos ayuda a encontrar las diferencias entre 2 archivos de nuestro proyecto, esto es til si por ejemplo cometimos un error y queremos comparar nuestro cdigo actual contra la ltima versin que no contena el error, de esta manera podemos ver que fu lo que cambiamos


Confirmaciones al remoto
Para enviar la ltima confirmacin de nuestro [root] utilizamos el comando [git push]
