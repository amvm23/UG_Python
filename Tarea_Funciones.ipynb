# Funciones en Python #


## Función ##
---
Una función nos permite definir un bloque de código reutilizable que se puede ejecutar muchas veces dentro de nuestro programa.  De las grandes ventajes de utilizar funciones es que reducimos la cantidad de líneas de código. Otras ventajas son: hace que el código sea más legible, facilita la depuración y limita los errores de escritura.

---

## Sintáxis ##
---
En Python, una definición de función tiene las siguientes características:

> 1. La palabra clave **def**
> 2. Un nombre de función
> 3. Paréntesis '()', y dentro de los paréntesis los parámetros de entrada
> 4. Dos puntos ':'
> 5. Algún bloque de código para ejecutar
> 6. Una sentencia de retorno (opcional)

---

## Ejemplo ##
---
>
>#esta es una función básica de suma
>def suma(a, b):
>  return a + b
>
>result = suma(1, 2)
>
>#result = 3
>

---

### Parámetros Posicionales y Parámetros por Nombre###

Cuando invocamos a una función en Python con diferentes parámetros o argumentos, los valores se asignan a los parámetros en el mismo orden en que se indican.  A esto le llamamos Parámetros Posicionales.  Ejemplo:
>
> def saludo(nombre, mensaje):
>
>      print("Hola {}, {}".format(nombre, mensaje))
>
---
> saludo("Jorge", "¿cómo estás?")
>
> **Hola Jorge, ¿cómo estás?**
---

Sin embargo, el orden se puede cambiar si llamamos a la función indicando el nombre de los parámetros.  A esto le llamamos Parámetros por Nombre

---
> saludo(mensaje="¿cómo estás?", nombre="Jorge")
>
> **Hola Jorge, ¿cómo estás?**
---

### Retorno de múltiples valores ###

Las funciones en Python nos proporcionan la capacidad de devolver no solo un resultado, podemos devolver múltiples resultados, simplemente separando los objetos por coma, incluso no tienen que ser del mismo tipo.  La forma en que recibimos estos valores es mediante una **Tupla**


>def suma_resta(a = 0, b = 0):
>
>     return a+b, a-b

---
>
>print(suma_resta(b=5,a=4))
>
>#(9, -1)

---

### Funciones Lambda o Anónimas ###

En Python, las funciones lambda son también conocidas como funciones anónimas porque se definen sin un nombre.  En vez de definirla con la palabra reservada **def**, utilizamos la palabra reservada **lambda**

---
>
> lambda parámetros: expresión
>
---

Las características principales de una función anónima:

- Son funciones que pueden definir cualquier número de parámetros pero una única expresión. Esta expresión es evaluada y devuelta.
- Se pueden usar en cualquier lugar en el que una función sea requerida.
- Estas funciones están restringidas al uso de una sola expresión.
- Se suelen usar en combinación con otras funciones, generalmente como argumentos de otra función

---
>
> cuadrado = lambda x: x ** 2
>
---

>
> print(cuadrado(3))
>
> #9
>

---

### Funciones como Objetos y Funciones como Parámetros ###

Python soporta el concepto de funciones de primera clase

>**Propiedades**
>
> ---
>
> 1. Una función es una instancia del tipo Objeto
> 2. Se puede guardar una función en una variable
> 3. Se puede pasar una función a otra función como parámetro
> 4. Se puede retonar una función como resultado de una función
> 5. Se puede guardar en estructura de datos como tablas hash , listas, tuplas, etc
>

```python
##Funciones como Objetos##

def suma(a,b):
    return a+b

adicion = suma

print(adicion(1,2))
```

```python
##Funciones como Argumentos a otras Funciones##

def resta(a,b):
    return a-b

def calculadora(a, b, suma, resta):
    S = suma(a,b)
    R = resta(a,b)
    return S, R

print(calculadora(4, 4, suma, resta))
```
