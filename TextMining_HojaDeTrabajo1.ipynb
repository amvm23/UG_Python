```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(pic, color):
    
    img = cv2.imread(pic) #BGR
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #RGB
    alto = img.shape[0]
    ancho = img.shape[1]
    lienzo = np.zeros((alto, ancho, 3))
    gris_ponderado = np.zeros((alto, ancho, 1))
    
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]

            #RGB
            red = pixel[0]
            green = pixel[1]
            blue = pixel[2]

            if color == 1:
                lienzo[i, j] = [0, 0, blue] #monocromo azul
            elif color == 2:
                lienzo[i, j] = [0, green, 0] #monocromo verde
            elif color == 3:
                lienzo[i, j] = [red, 0, 0] #monocromo verde
            elif color == 10:
                lienzo[i, j] = [red, green, 0] #rojo y verde
            elif color == 20:
                lienzo[i, j] = [0, green, blue] #verde y azul
            elif color == 30:
                lienzo[i, j] = [red, 0, blue] #verde y azul
            elif color == 50:
                gris_ponderado[i, j] = int(0.25*blue + 0.50*green + 0.25*red) #gris ponderado 
            else:
                lienzo[i, j] = [red, green, blue] #full

    if color == 50:
        cv2.imwrite("monocromo.jpg", gris_ponderado)
    else:
        cv2.imwrite("monocromo.jpg", lienzo)
    
    monocromo = cv2.imread("monocromo.jpg")
    plt.imshow(monocromo)
    plt.show()
    
    return monocromo
```

```python
def show_image_mono3D(pic1, pic2, pic3):
    #pic1="perro_salida_gray_rojo.jpg"
    #pic2="perro_salida_gray_verde.jpg"
    #pic3="perro_salida_gray_azul.jpg"
    img1 = cv2.imread(pic1) #BGR
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) #RGB
    img2 = cv2.imread(pic2) #BGR
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB) #RGB
    img3 = cv2.imread(pic3) #BGR
    img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB) #RGB


    alto = img1.shape[0]
    ancho = img1.shape[1]

    lienzo = np.zeros((alto, ancho, 3))
    
    for i in range(0, alto):
            for j in range(0, ancho):
                pixel1 = img1[i, j]
                pixel2 = img2[i, j]
                pixel3 = img3[i, j]

                #RGB
                red = pixel1[0]
                green = pixel2[1]
                blue = pixel3[2]

                lienzo[i, j] = [red, green, blue] #color original
                

    cv2.imwrite("imagen_original.jpg", lienzo)
    original = cv2.imread("imagen_original.jpg")
    plt.imshow(original)
    #plt.hist(lienzo_rojo,bins=20)
    
    plt.show()

    return 0
```

```python
def show_escala_grises(pic):

    #pic="paris.jpg"
    img = cv2.imread(pic) #BGR
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #RGB

    alto = img.shape[0]
    ancho = img.shape[1]

    lienzo_rojo = np.zeros((alto, ancho, 1))
    lienzo_verde = np.zeros((alto, ancho, 1))
    lienzo_azul = np.zeros((alto, ancho, 1))

    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]

            #RGB
            red = pixel[0]
            green = pixel[1]
            blue = pixel[2]

            lienzo_rojo[i, j] = red
            lienzo_verde[i, j] = green
            lienzo_azul[i, j] = blue

    cv2.imwrite("gris_rojo.jpg", lienzo_rojo)
    cv2.imwrite("gris_verde.jpg", lienzo_verde)
    cv2.imwrite("gris_azul.jpg", lienzo_azul)

    rojo = cv2.imread("gris_rojo.jpg")
    verde = cv2.imread("gris_verde.jpg")
    azul = cv2.imread("gris_azul.jpg")

    plt.imshow(rojo)
    plt.show()

    plt.imshow(verde)
    plt.show()

    plt.imshow(azul)
    plt.show()

    plt.imshow(img)
    plt.show()
    
    return 0
        

```

# Modelo HSV #
---
**El espacio HSV (o HSB)** define los colores en función de tres factores

### Matiz
---
*Es lo que común -y erróneamente- llamamos color: la longitud de onda dominante. *

### Saturación
---
*Mide la pureza de un color, es decir la ausencia de otros colores: un color con el 100% de saturación se conseguirá tan solo con su matiz monocromático o espectral.*

### Valor o Brillo 
---
*Hace referencia a la intensidad de la luz, es decir, la cantidad de luz que hay en ese color, desde un 0% (negro) al 100%.*
